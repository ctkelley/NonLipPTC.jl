This example is the integral equation
$$
u(x) - \int_0^1 \sin(2 \pi (x - y) u_+^{1/2} (y) \  dy = f(x).
$$
We build $f$ so that the exact solution is $u^*(x) = x^{3/2}\cos(10x)$. This choice makes the iteration 
converge slowly enough so we can see the differences in the parameter choices.

Define the nonlinear integral operator
$$ 
K(u)(x) = \int_0^1 \sin(2 \pi (x - y) u^{1/2}_+ (y) \ dy
$$
Then we can write
$$
F(u) = u - K(u) - f
$$
We show that $F$ is strongly monotone by showing that
$$
(K(u) - K(v), u-v) \equiv \int_0^1 (K(u)(x) - K(v)(x) ) (u(x)-v(x) ) \ dx = 0
$$ 
for all $u, v \in L^2$. We got this proof from Gemini. Note that
$$
K(u) - K(v) = \int_0^1 \sin(2 \pi (x - y) (u^{1/2}(y) - v^{1/2})(y) \  dy
$$
So, letting $g(y) = ^{1/2}(y) - v^{1/2}(y)$ and $d(x) = u(x) - v(x)$, we can write
$$
(K(u) - K(v), u-v) = \int_0^1 \int_0^1 \sin(2 \pi (x-y) g(y) d(x) \ dx  \ dy
$$
For any function $f(x,y)$ we know that
$$
\int_0^1 \int_0^1 f(x,y) \ dx \ dy = \int_0^1 \int_0^1 f(y,x) \ dx \ dy 
$$
we have
$$
\begin{array}{ll}
(K(u) - K(v), u-v) &= \int_0^1 \int_0^1 \sin(2 \pi (x-y) g(y) d(x) \ dx \ dy\\
& = - \int_0^1 \int_0^1 \sin(2 \pi (y-x) g(y) d(x) \ dx \ dy\\
& = - \int_0^1 \int_0^1 \sin(2 \pi (x - y) g(x) d(y) \ dx \ dy \\
& = - (K(u) - K(v), u-v) 
\end{array}
$$
and hence $(K(u) - K(v), u-v) = 0$
